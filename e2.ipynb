```matlab
a = 0; b = 1; n = 3;
precision = 0.0001;
u = a:precision:b;

xt = [1,1,1,0];
yt = [0,0,0,1];

cxt = zeros(1, length(xt));
cyt = zeros(1, length(yt));

for i = 0:length(xt)-1
    for j = 0:i
        cxt(i+1) = cxt(i+1) + xt(j+1) * nchoosek(n-j, i-j)/nchoosek(n,i);
        cyt(i+1) = cyt(i+1) + yt(j+1) * nchoosek(n-j, i-j)/nchoosek(n,i);
    end
end

p_x = zeros(1, length(u));
p_y = zeros(1, length(u));

for i = 1:length(u)
    [t_x, t_y] = de_casteljau(n, cxt, cyt, u(i));
    p_x(i) = t_x(n+1, n+1); 
    p_y(i) = t_y(n+1, n+1);
end

plot(cxt, cyt, '-O'); hold on;
plot(p_x, p_y, 'linewidth',3); hold on;
```
