```matlab
a = 0; b = 7; n = 5;
precision = 0.0001;
u = a:precision:b;
r_u = (u-a)./(b-a);
x_p = [0, 2, 3, 5, 5.5, 7];
y_p = [0, 2, 1, 3, 0.5, 1];
p_x = zeros(1, length(u));
p_y = zeros(1, length(u));
for i = 1:length(u)
    [t_x, t_y] = de_casteljau(n, x_p, y_p, r_u(i));
    p_x(i) = t_x(n+1, n+1); 
    p_y(i) = t_y(n+1, n+1);
end
```

```matlab
plot(x_p, y_p, '-O'); hold on
plot(p_x, p_y, 'linewidth',3); hold on;
```
