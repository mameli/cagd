```matlab
u = linspace(0, 10, 100);
v = linspace(0, 10, 100);
[uu, vv] = meshgrid(u);
x = uu;
y = vv;
z = sin(vv);
surf(x, y, z);
```

```matlab
k_u = 3; k_v = 3;
knots_u = [zeros(1, k_u), ones(1, k_u)];
knots_v = [zeros(1, k_v), ones(1, k_v)];
tab = 0: 0.05 : 1;
B_u = spcol(knots_u, k_u, tab);
B_v = spcol(knots_v, k_v, tab);
for i = 1:length(knots_u) - k_u
    for j = 1:length(knots_v) - k_v
        X = B_u(:,i)*B_v(:,j).';
        surf(tab, tab, X, 'FaceAlpha', 0.8); shading flat;
        s.EdgeColor = 'none';
        hold on;
    end
end
```

```matlab
k_u = 3; k_v = 3;
b_x = [1 2 3; 1 2 4; 1 2 4];
b_y = [4 6 4; 3 5 2; 3 2 1];
b_z = [2 2 2; 2 4 2; 2 5 3];
plot3(b_x, b_y, b_z, 'bo--', 'linewidth', 2); hold on;
plot3(b_x', b_y', b_z', 'bo--', 'linewidth', 2); 
grid on; axis tight; axis equal;
knots_u = [zeros(1, k_u), ones(1, k_v)];
knots_v = [zeros(1, k_u), ones(1, k_v)];
tab = 0: 0.05 : 1;
B_u = spcol(knots_u, k_u, tab);
B_v = spcol(knots_v, k_v, tab);
X_x = B_u*b_x*B_v.';
X_y = B_u*b_y*B_v.';
X_z = B_u*b_z*B_v.';
surf(X_x, X_y, X_z, 'FaceAlpha', 0.8); 
shading flat; s.EdgeColor = 'none';
```
